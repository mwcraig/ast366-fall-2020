## Planning considerations -- summary

### Where to find object

+ Where in the sky is it -- need to know where to point telescope 
+ Coordinates of galaxy
+ What constellation -- to get general location

### What can we change about view

+ Distance from us to what we observing -- may need to know to adjust focus
    - Don't need to adjust focus because of distance to object
    - Do need to focus because of changes in size of telescope (due to temperature) and changes in seeing conditions (how turbulent is the atmosphere)
    
+ Size/width -- so that we can fill field of view with object
    - Cannot adjust field of view.
    - Do need to check whether the object is too small to resolve or too big to fit in frame
    
+ Take bias/dark/flat calibration image -- remove noise, know physical capabilities of telescope.

### Can we see the thing we want to see?

+ Brightness/magnitude -- can the telescope observe it? Is it too bright to measure?
+ Exposure time -- may be related to wavelength, affects quality/detail
+ Phase of the moon and where it is relative to the thing you are observing -- affects quality of image maybe

### How do we make things happen?

+ Commands/programming laguage to be able to control telescope to align it.

### Other

+ Compensate for drift (rotation of earth) manually
    - Up to 5 minute exposures with no drifting at all
    - For longer exposure times, take multiple pictures and combine.
    
+ Timing -- when is object visible?
